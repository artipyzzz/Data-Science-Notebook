# Additional methods

- .count() - метод списка который возвращает кол-во указанных элементов в списке
- .insert() - метод который добавляет элемент в список в указанный индекс
- .pop() - методы который удаляет элемент с указанным индексом и возвращает этот элемент
- range() - встроенная функция которая создает очередность целых чисел
- len() - встроенная функция которая возвращает длинну объекта
- .sort() / sorted() - встроенные методы которые сортируют элементы в листе

### .inser(index, new_value)

```python
# создадим список из 5 целых чисел
my_list = list(range(5))
print(my_list)
# Добавим в список элемент используя метод insert
my_list.insert(2, 100)
print('Список после добавления нового элемента под индекс 2', my_list)
# как видно из вывода элемент становится под указанный индекс все последующие элементы сдвигаются на один индекс вправо
```

### .pop(index)
Если методу pop() не передавать в скобки ни какие аргументы то по умолчанию будет установлен индекс -1. То есть будет удален последний элемент списка

```python
# создадим список из 5 целых числе
my_list_2 = list(range(5))
print(my_list_2)
# удалим один элемент под индексом 2
second_element = my_list_2.pop(2)
print('second element:', second_element)
print('my_list_2:', my_list_2)
# как видно из вывода элемент с индексом 2 был удален из списка и сохранен в переменную second_element
# сохранение произошло потому что метод pop возвращает элемент после удаление его из списка
```

### range(start, end, step)

Встроенный метод range() создает последовательность целых чисел где
- start - целое число начала отсчета (если значение не указанно по умолчанию 0)
- end - целое число конца отсчета (он не учитывается при создании последовательности)
- step - шаг перечисления (если не указан по умолчанию 1)

```python
# создадим обычную последовательность из 5 элементов
# чтобы увидеть всю последовательность следует объект класса range() перевести в тип данных список
print(list(range(5)))

# попробуем создать список элементов от 3 до 7
print(list(range(3, 8)))
# прошу обратить внимание что верхня граница range не учитывается и потому для того чтобы в списке значения были от 3 до 7 включая 7 верхняя граница была указана 8

# попробуем создать список из четных чисел используя метод range()
print(list(range(2, 11, 2)))
# для создания подобного списка был указан шаг

# шаг может быть и отрицательный но при это начальная граница должна быть больше конечной
print(list(range(10, 0, -1)))
```

### len(object)

Встроенная функция len() возвращает длинну объекта в случае если ее можно вычислить. Объекты длинну которых можно получить:
- string - строка
- tuple - кортеж
- dictionary - словарь
- list - список
- set - множество

```python
from numpy import random

# создадим список из случайного кол-во целых чисел длинной от 50 до 70 элементов
my_list_3 = list(range(random.randint(50, 70)))
# метод len() позволит нам понять сколько элементов содержиться в списке
print(len(my_list_3))
```

### slicing list

У списков можно делать срезы, другими словами получать только тот отрезок списка который мы укажем используя индексы

```python
# создадим список из 10 элементов, каждое число которого будет равно своему индексу
my_list_4 = list(range(10))
print(my_list_4)

# для того чтобы сделать срез следует воспользоваться следующей структурой
# предположим что нам требуется получить элементы с индексами от 1 до 3 включительно
# важно учитывать что в срезах второй порог не учитывается
slice_1_3 = my_list_4[1:4]
print(slice_1_3)

# также с помощью срезов мы можем выбрать элементы с четным или нечетным индексом добавив дополнительный элемент
slice_even_index = my_list_4[::2]
print(slice_even_index)

# не четный индекс. Обратите внимание что был указан только один параметр среза (начальный индекс))
slice_odd_index = my_list_4[1::2]
print(slice_odd_index)

# Получается что мы можем указать только некоторые аргументы среза
slice_1_to_end = my_list_4[1:]
print(slice_1_to_end)
slice_0_to_4 = my_list_4[:5]
print(slice_0_to_4)
```

### .count(value_in_list)

.count() метод списка который предоставляет кол-во указанных элементов в списке

```python
# создадим список из 10 элементов каждый элемент будет случайно выбираться из списка [0, 1, 2] по средствам модуля random
my_list_5 = [int(random.choice([0,1,2])) for i in range(10)]
print(my_list_5)

# мы знаем что в списке есть три уникальных элемента 0, 1 или 2 но не знаем в каком кол-ве
# в таком случае нам поможем метод .count()
for i in [0,1,2]:
    print(f"В списке my_list_5 элемент {i} встречается {my_list_5.count(i)} раз.")
```

### sorting list .sort() | sorted()

Существует два метод сортировки элементов в списке
- .sort() - данный метод применяется от объекта списка. Данный метод сортирует элементы в самом списке изменяя его.
- sorted() - данный метод применяет на объект списка. Данный возвращает новый отсортированный список.

```python
# создадим список случайных чисел от 0 до 5 длинной 10
my_list_6 = [int(random.choice(list(range(6)))) for i in range(10)]
print(my_list_6)

# для начала применим метод sorted
my_list_6_sorted = sorted(my_list_6)
print('Новый отсортированный список методом sorted', my_list_6_sorted)
print('Список который сортировали', my_list_6)
# как видно из результата новый списко стал отсортированным
# а список элементы которого сортировали остался не изменным

# теперь попробуем отсортировать список методом .sort()
my_list_6_sort = my_list_6.sort()
print('Новый отсортированный список методом .sort()', my_list_6_sort)
print('Список который сортировали', my_list_6)
# по результату можно заметить что новый список ничего не содержит, это произошло потому что метод .sort() ничего не возвращает а изменяет исходных список
# поэтому в строке "Список который сортировали" сразу отсортирован

# для того чтобы произвести сортировку по убывающей используется параметр reverse который является булевым
my_list_6.sort(reverse=True)
print('my_list_6.sort(reverse=True)', my_list_6)
my_list_6_sorted_reverse = sorted(my_list_6, reverse=True)
print('my_list_6_sorted_reverse', my_list_6_sorted_reverse)
```
