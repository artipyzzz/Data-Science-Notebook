## Столбчатая диаграмма с ошибкой

Данные, которые вам понадобятся для воссоздания этой диаграммы, находятся в переменных
- past_years_averages
- years

Сохраните воссозданную диаграмму в файле под названием *`my_bar_chart.png`*

```python
import matplotlib.pyplot as plt

past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]

# сначала воссоздайте фигуру размерами 10 на 8
plt.figure(figsize=(10,8))

plt.bar(range(len(past_years_averages)), past_years_averages, yerr = error, capsize = 5)
plt.axis([-.5, 6.5, 70, 95])

ax = plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

plt.title('Final Exam Averages')
plt.xlabel('Year')
plt.ylabel('Test average')

plt.savefig('my_bar_chart.png')

plt.show()
```

# Side-by-Side bar

Теперь давайте рассмотрим диаграмму ниже

![](../img/p_1.png)

Она отображает разницу в средних результатах тестов между учениками, которые учились в двух разных средних школах до поступления в вашу старшую школу.

Данные, которые вам понадобятся для воссоздания этой диаграммы, находятся в переменных:
- middle_school_a
- middle_school_b
- unit_topics

Сохраните воссозданную диаграмму в файле с именем *`my_side_by_side.png`*

```python
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]

# используя функцию create_x создайте переменные school_a_x, school_b_x
school_a_x = create_x(2, 0.8, 1, 5)
school_b_x = create_x(2, 0.8, 2, 5)

# создайте фигуру размером 10 на 8
plt.figure(figsize=(10,8))

# создайте набор осей в переменной ax
ax = plt.subplot()

# постройте на одном графике ряд столбцов, представляющих middle_school_a и middle_school_b
plt.bar(school_a_x, middle_school_a)
plt.bar(school_b_x, middle_school_b)

# создайте новый список значений x под названием middle_x, которые являются значениями в середине school_a_x и school_b_x. Это то место, где мы разместим x-ticks
middle_x = [(a + b)/2 for a,b in zip(school_a_x,school_b_x)]
ax.set_xticks(middle_x)
ax.set_xticklabels(unit_topics)

plt.legend(['Middle School A', 'Middle School B'])
plt.title('Test Averages on Different Units')
plt.xlabel('units')
plt.ylabel('test average')

plt.savefig('my_side_by_side.png')

plt.show()
```

# Complex bar

Теперь давайте рассмотрим диаграмму ниже

![](../img/p_2.png)

Эта диаграмма отображает распределение студентов, получивших оценки A,B,C,D и F по каждому разделу

Данные которые вам понадобятся для воссоздания этой диаграммы, сохраненны в переменных:
- As
- Bs
- Cs
- Ds
- Fs
- unit_topics

Сохраните воссозданную диаграмму в файле с именем *`my_stacked_bar.png`*

```python
import numpy as np

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)

# столбцы Bs будут располагаться поверх As, но на какой высоте будут начинаться остальные ? Нижняя часть полос, представляющих Cs будут на высоте As и Bs. Мы можем сделать это в Numpy с помощью функции np.add(). c_bottom, начальные высоты для Cs. Рассчитайте значения для d_bottom, f_bottom.
c_bottom = np.add(As, Bs)
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)

# создайте фигуру размером 10 на 8
plt.figure(figsize=(10,8))

# выведите график. Дайте каждому из них соответствующий bottom список, который наложит их друг на друга
plt.bar(range(len(unit_topics)), As)
plt.bar(range(len(unit_topics)), Bs, bottom=As)
plt.bar(range(len(unit_topics)), Cs, bottom=c_bottom)
plt.bar(range(len(unit_topics)), Ds, bottom=d_bottom)
plt.bar(range(len(unit_topics)), Fs, bottom=f_bottom)

# создайте набор осей
ax = plt.subplot()

# добавьте наименования меток по оси Х
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)

# дайте графику название Grade Distribution
plt.title('Grade Distribution')

# добавьте обозначение осей
plt.xlabel('Unit')
plt.ylabel('Number of Students')

# сохраните рисунок
plt.savefig('my_stacked_bar.png')

plt.show()
```

# Line with Filled Error

Теперь давайте рассмотрим диаграмму

![](../img/p_5.png)

Она отображает связь между временем, которое, по словам студентов, они потратили на учебу, и баллами, которые они получили на выпускных экзаменах.

Данные которые вам понадобятся для воссоздания этой диаграммы сохранены в переменные:
- hours_reported
- exam_scores

Сохраните свой график в файл с именем my_line_graph.png

```python
hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]

# создайте фигуру размером 10 на 8
plt.figure(figsize=(10,8))

# постройте график exam_scores на оси X и hours_reported на оси Y, при этом linewidth равно 2
plt.plot(exam_scores, hours_reported, linewidth=2)

# предположим что погрешность в отчетах студентов о часах учебы составляет 20%. Создайте список hours_lower_bound содержащий элементы hours_reported минус 20%, и список hours_upper_biund, содержащий элементы hours_reported плюс 20%.
hours_lower_bound = [i - i*.2 for i in hours_reported]
hours_upper_bound = [i + i*.2 for i in hours_reported]

# создайте линию заштризованную линию погрешности
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha = .2)

# дайте название графику
plt.title('Time spent studying vs final exam scores')
plt.xlabel('Score')
plt.ylabel('Hours Studying (self-reported)')

# сохраните рисунок в файл с именем my_line_graph.png
plt.savefig('my_line_graph.png')

plt.show()
```
