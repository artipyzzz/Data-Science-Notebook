````python
# импортируем библиотеки
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# загружаем данные в переменную
df = pd.read_csv('../files/mushroom_data.csv')

# выводим пару строк данных
print(df.head())

# создаем переменную с наименованием колонок данных
columns = df.columns.tolist()
````

# Построение столбчатых диаграмм

В этом наборе данных 23 переменные (по одной на каждый столбец). Построение графика для каждой из них по отдельности было бы утомительно; к счастью, вы будете использовать циклы.

Именно для этого ранее вам требовалась создать переменную со списком наименований колонок набора данных.

- Создайте цикл который проходит по каждому элементу списка *`columns`*
- Выведите на экран каждый элемент

````python
for column in columns:
    print(column)
````

Отлично цикл работает как надо. Теперь можно приступать к построение графиков с использованием метода *`.countplot()`* из библиотеки *`seaborn`*. Выполните следующие шаги:
- Заведите метод *`.countplot()`* в цикл *`for`* который был создан ранее
- Используя структуру вызова данных по имени колонки pandas постройте графики дла каждой колонки

В конце каждой иттерации добавьте строки кода:

```python
plt.show()
plt.clf()
```

````python
for column in columns:
    sns.countplot(df[column])
    # строки кода для улучшения читаемости
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    plt.show()
    plt.clf()
````
