# Линейные графики Sublime Limes

В этом проекте вы будете выступать в качестве аналитика данных для интрнет-магазина лаймов. Люди по всему миру могут использовать этот продукт, чтобы получать самые свежие и качественные лаймы с доставкой на дом. Один из менеджеров по продукту хотел бы получить представление о клиентах и их привычках в продажах. Используя Matplotlib, вы создадите несколько различных линейных графиков для эффективном передачи этой информации

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

```python
# создайте фигуру размером 12 на 8 (ширина на длинну)
plt.figure(figsize=(12 ,8))

# график должен содержать два подграфика в одну строку
ax1 = plt.subplot(1,2,1)

# На левом подграфике мы изобразим в виде линии общее количество посещений страниц за последний год. Сначала создадим список значений x, который равен range(len(months)). Сохраним его в переменной с именем x_values.
x_values = range(len(months))

# постройте график общего кол-ва посещений страниц в зависимости от значений x_values в виде линии
# обозначьте каждый маркер именем месяца
plt.plot(x_values, visits_per_month, marker='o')
plt.title('Visits - Months')

# добавьте наименования осей
ax1.set_xlabel('Months')
ax1.set_ylabel('Visits')

# Добавьте тикеры по оси X и обозначьте значения именами месяца
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

# второй подграфик
ax2 = plt.subplot(1,2,2)

# на данном подграфике мы построим три линии относительно тех же значений X. Значения X для всех трех линий будут соответствовать месяцам, поэтому мы можем использовать список, который x_values мы использовали для крайнего графика.

# На одном участке создать три линии
# Кол-во проданных лаймов
# Кол-во проданных персидких лаймов
# Кол-во проданых кровавых лаймов
x_values = range(len(months))

plt.plot(x_values, key_limes_per_month, marker = 'o')
plt.plot(x_values, persian_limes_per_month, marker = 'o')
plt.plot(x_values, blood_limes_per_month, marker = 'o')
plt.title('Sales - Months')
# добавьте легенду
plt.legend(['Key', 'Persian', 'Bloody Red'])

ax2.set_xticks(x_values)
ax2.set_xticklabels(months)

# сохраните график
plt.savefig('Page Visits Limes Sales.png')
plt.show()
```
