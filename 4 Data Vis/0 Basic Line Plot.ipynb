# Basic Line

Линейные графики полезны для визуализации изменения переменной с течением времени.

Некоторые возможные данные, которые можно отобразить с помощью линейного графика:
- средние цены на бензин за последнее десятилетие
- вес человека за последние пару месяцев
- средняя температруа вдоль линии долготы на разных широтах

Используя методы Matplotlib, следующий код создаст простой линейный график `.plot()` и отобразит его с помощью `.show()`

```python
x_values = [0, 1, 2, 3, 4]
y_values = [0, 1, 4, 9, 16]
plt.plot(x_values, y_values)
plt.show()

```

- x_values - это переменная, содержащая список значений X для каждой точки на нашем линейном графике
- y_values - это переменная, содержащая список значений Y для каждой точки на нашей линейном графике
- plt - это имя, которое мы дали модулю matplotlib, импортированному в верхней части кода
- plt.plot(x_values, y_values) - создаст линейный график
- plt.show() - на самом деле отобразит график

Наш график будет выглядеть так:

![](./img/0_1.png)

Давайте попрактикуемся в построении линий

````python
# Мы собираемся сделать простой график, представляющий чьи-то расходы на обед за последнюю неделю. Сначала определим два списка, days и money_spent, которые содержат следующие целые числа
import matplotlib.pyplot as plt

days = list(range(7))
money_spent = [10,12,12,10,14,22,24]
````

````python
# постройте график days по оси X и money_spent по оси Y, используя plt.plot()
plt.plot(days, money_spent)
plt.show()
````

Matplotlib автоматически разместит две линии на одних и тех же осях и присвоит им разные цвета, если вы вызовете ее `plt.plot()` дважды.

# Axis and Lables

Иногда может быть полезно увеличить или уменьшить масштаб графика, особенно если есть какая-то деталь, которую мы хоти рассмотреть. Для увеличения масштаба мы можем использовать `plt.axis()`. Мы используем `plt.axis()`, передавая ему список в качестве входных данных. Этот список должен содержать:
- минимальное отображаемое значеие X
- максимальное отображаемое значение X
- минимальное отображаемое значеие Y

```python
x = [0, 1, 2, 3, 4]

plt.show()
```

# Subplots

Иногда мы хоти отобразить две линии рядом, а не в одном наборе осей X и Y. Когда у нас есть несколько осей на одном рисунке, мы называем каждый набор осей subplot-ами. Рисунок или объект, который содержит все subplot-ы, называется фигурой.

У нас может быть много разных subplot-ов на одном рисунке, и мы можем выложить их множеством разных способов. Мы можем думать о наших макетах как о строках и столбцах subplot-ов. Например, на следующем рисунке шесть subplot-ов разделены на 2 строки и 3 столбца

![](./img/0_7.png)

Мы можем создать subplot используя метод *`.subplot()`*

Команде *`plt.subplot()`* необходимо передать три аргумента:
- количество рядов
- количество столбцов
- индекс subplot-а, который мы хотим создать

Например, команда *`plt.subplot(2,3,4)`* создаст "Subplot 4" из рисунка выше.

Любые *`plt.plot()`* последующие действия *`plt.subplot()`* создаст линейный график в указанном подграфике. Например:

```python
# Data sets
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

# First Subplot
plt.subplot(1, 2, 1)
plt.plot(x, y, color='green')
plt.title('First Subplot')

# Second Subplot
plt.subplot(1, 2, 2)
plt.plot(x, y, color='steelblue')
plt.title('Second Subplot')

# Display both subplots
plt.show()
```

В результате получим рисунок с двумя графиками, расположенными следующим образом:

![](./img/0_8.png)

````python
# По уже созданным спискам months, temperature и flight_to_hawaii. Используйте plt.subplot команду, постройте temperature - months в левом поле рисунка, который имеет 1 строку с 2 столбцами
months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]

plt.subplot(1,2,1)
plt.plot(months, temperature)

# график flights_to_hawaii - temperature на том же рисунке, справ от вашего первого графика. Добавьте параметр "о" в конец вашего вызова, чтобы plt.plot() превратить график в диаграмму рассеяния
plt.subplot(1,2,2)
plt.plot(flights_to_hawaii, temperature, "o")

plt.show()
````

# Subplot 2

Иногда, когда мы объединяем несколько подсюжетов, некоторые элементы могут перекрываться и делать рисунок нечитаемым:

![](./img/0_9.png)

Мы можем настроить интервал между напиши подграфиками, тчобы убедиться, что созданная нами фигура видна и понятна. Для этого мы используем *`plt.subplots_adjust()`* команду. Данная команда имеет несколько ключевых аргументов, которые могут перемещать ваши графики внутри фигуры.

- left - левое поле, по умолчанию 0.125. Вы можете увеличить это число, тчобы освободить место для метки оси Y
- right - правое поле, по умолчанию 0.9. Вы можете увеличить его, чтобы освободить место для рисунка, или уменьшить, чтобы освободить место для легенды
- bottom - нижнее поле, по умолчанию 0.1. Вы можете увеличить его, чтобы освободить место для меток делений или метки оси X
- top - верхнее поле, по умолчанию 0.9
- wspace - горизонтальное расстояние между соседними участками, по умолчанию 0.2
- hspace - вертикальное расстояние между соседними подучатсками, по умолчанию 0.2

Например, если бы мы добавляли пространство внизу графика, изменяя нижнее поле до 0.2 (вместо значения по умолчанию 0.1), мы бы использовали команду

*`plt.subplots_adjust(bottom=0.2)`*

Мы также можем использовать несколько ключевых аргументов, если нам нужно настроить несколько полей. Например, мы могли бы настроить top и hspace:

*`plt.subplots_adjust(top=0.95, hspace=0.25)`*

Давайте воспользуемся wspace для исправления рисунка выше:

```python
# Left Plot
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

# Right Plot
plt.subplot(1, 2, 2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

# Subplot Adjust
plt.subplots_adjust(wspace=0.35)

plt.show()
```

Это даст нам рисунок с лучшей компановкой

![](./img/0_10.png)

````python
# Давайте соберем фигуру, котоаря имеется два ряда подграфиков. Она должна иметь
# один подсюжет в верхнем ряду
# два подсюжета в нижнем ряду
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

plt.subplot(2,1,1)
# график streight_line - x
plt.plot(x, straight_line)

plt.subplot(2,2,3)
plt.plot(x, parabola)
plt.subplot(2,2,4)
plt.plot(x, cubic)

# увеличьте расстояие между горизонтальными подграфиками до 0.35, а нижнее поле до 0.2
plt.subplots_adjust(wspace=0.35, bottom=0.2)

plt.show()
````

# Legends

Если на одном графике имеется несколько линий, мы можем пометить их с помощью команды *`plt.legend()`*

Метод *`legend`* принимает список с метками для отображения. Так, например, мы можем вызвать:

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
plt.legend(['parabola', 'cubic'])
plt.show()
```

Который отобразит легенду на нашем графике, маркируя каждую линию

![](./img/0_11.png)

*`plt.legend()`* также может принимать ключевой аргумент *`loc`*, который будет располагать легенду на рисунке.

Принимаются следующие щначения позиций *`loc`*

```text
0	лучший
1	вверху справа
2	верхний левый
3	нижний левый
4	нижний правый
5	верно
6	центр слева
7	центр справа
8	нижний центр
9	верхний центр
10	центр
```

**Примечание:** Если вы решите не устанавливать значение *`loc`*, по умолчанию будет выбрано "лучшее" местополжение

Например, мы можем вызвать *`plt.legend()`* и *`loc`* установить 6:

```python
plt.legend(['parabola', 'cubic'], loc=6)
plt.show()
```

Что переместит легенду в левую часть графика:

![](./img/0_12.png)

# Modify Ticks

Во всех наших предыдущих упражнениях наши команды начинались с *`plt.`*. Чтобы изменить отметки, нам придется попробовать что-то немного другое.

Поскольку нашу сюжеты могут иметь несколько подсюжетов, нам нужно указать, какой из них мы хотим изменить. ЧТобы сделать это, мы вызываем *`plt.subplot()`* по другому

```python
ax = plt.subplot(1, 1, 1)
```

*`ax`* является объектом осей, и он позволяет нам изменять оси, принадлежащие определенному подучастку. Даже если у нас есть только один подучасток, когда мы хоти изменить отметки, нам нужно будет начать с вызова *`ax = plt.subplot(1,1,1)`* или *`ax = plt.subplot()`* для того, чтобы получить наш объект осей.

Предположим, мы хотим установить наши крестики на значения 1, 2 и 4. Мы бы использовали следующий код

```python
ax = plt.subplot()
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
ax.set_xticks([1, 2, 4])
```

Наш результат будет выглядеть так

![](./img/0_13.png)

Мы также можем изменить y-метки с помощью *`ax.set_yticks()`*.

Когда мы меняем x-метки, их метки автоматически меняются в соответствии с ними. Но если нам нужны специальные метки (например, строки), мы можем использовать команду *`ax.set_xticklabels()`* или *`ax.set_yticklabels()`*. Например, мы можем захотеть иметь ось Y с метками на 0.1, 0.6 и 0.8, но пометить их как 10%, 60% и 80% соответственно. Для этого мы используем следующие команды

```python
ax = plt.subplot()
plt.plot([1, 3, 3.5], [0.1, 0.6, 0.8], 'o')
ax.set_yticks([0.1, 0.6, 0.8])
ax.set_yticklabels(['10%', '60%', '80%'])
```

Это приведет к следующей маркировке оси Y
