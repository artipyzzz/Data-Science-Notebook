## Введение в столбчатые диаграммы

Теперь, когда мы увидели, как выглядит столбчатая диаграмма, давайте рассмотрим более подробно, как и когда мы ее используем. Столбчатые диаграммы обычно используются для визуализации относительных частот категорий в пределах переменной. Поэтому они в первую очередь полезны для визуального обобщения категорийных переменных а не колличественных. Если вы можете организовать переменную в отдельные категории, столбчатая диаграмма часто является отличным вариантом. Если нет, вам, возможно придется прибегнуть к другому визуальному инструменту.

После того, как у вас есть отдельные категории, лучше всего использовать столбчатую диаграмму для отображения различных значений каждой категории. Мы также можем сравнивать средние значения, но мы бы рекомендовали использовать для этого диаграмму side-by-side, потому что они дают полную картину.

### Вертикальные и горизонтальные столбчатые диаграммы

В столбчатой диаграмме у нас есть две оси. Одна ось представляет дискретные значения внутри категориальной переменной, которую мы сравниваем, в то время как другая ось измеряет кол-во каждого из этих значений. В зависимости от того, как мы ориентируем эти оси, у нас может быть либо *`горизонтальная столбчатая диаграмме`*, либо *`вертикальная стобчатая диаграама`*. Давайте рассмотрим каждую из них

![](./img/2_2.png)

### Столбчатые диаграммы и гистограммы

Наконец, нам осталось рассмотреть еще одну вещь, прежде чем мы приступим к кодированию наших собственных диаграмм. Если вы когла-либо сталкивались с гистограммами, вы могли заметить, что столбчатые диаграммы и гистограммы выглядят почти одинаково. Однако вот основные различия между ними:
- Столбчатые диаграммы используются для категориальных переменных, а гистограммы - для количественных данных
- Гистограммы всегда должны быть расположены в определенном порядке, поскльку они представляют собой диапазон числовых значений. Для столбчатой диаграммы каждый столбец представляет частоты переменных категории в пределах категории. Если переменная не является порядковой, столбцы могут быть расположены в любом порядке.

![](./img/2_3.png)

````python
# Ниже мы написали код для создания столбчатой диаграммы и гистограммы с использование данных о супергероях из файла ./files/heroes_information.csv

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# загружаем данные из файла
heroes_data = pd.read_csv('./files/heroes_information.csv')

# очищаем данные
heroes_data_cleaned = heroes_data[heroes_data.Alignment != '-']

# создаем под график который будет располагаться слева
# и будет являться столбчатой диаграммой
plt.subplot(1,2,1)
sns.countplot(heroes_data_cleaned['Alignment'])

# создаем под график который будет располагаться справа
# и будет является гистограммой
plt.subplot(1,2,2)
sns.histplot(heroes_data_cleaned['Height'])
plt.xlim(0,500)

plt.show()
````

# Область столбчатой диаграммы

Давайте рассмотрим одну из важнейших концепций столбчатых диаграмм.

Столбики столбчатой диаграммы имеют несколько ключевых особенностей
- их длина пропорциональна кол-ву, которое они представляют
- ширина каждого столбца на гистограмме одинакова, то есть площадь каждого столбца также пропорциональна кол-ву представляемых им значений

Эти особенности делают столбчают диаграмму чрезвычайно надежным средством представления категориальных данных.

Для любой диаграммы, например столбчатой диаграммы, области, которые мы используем для представления значений, всегда должны быть эквивалентны относительным размерам значений, которые они представляют. В противном случае читатели могут быть введны в заблуждение и потенциально идентифицировать закономерности, который на самом деле нет.

# Круговая диаграмма

До сих пор в этом уроке мы рассматривали столбчатые диаграммы как способ прдеставления ктегориальных данных. Однако столбчатые диаграммы - не единственный вариант. Часто вы будете сталкиваться с *`круговыми диаграммами`* как способом отображения категориальных данных. Хотя столбчатые диаграммы обычно используются для визуализации таблицы частот, круговые диаграммы являются альтернативой, когда вы хотите визуализировать таблица пропорций.

Круговые диаграммы состоят из срезовЮ которые обхединены в полный круг. Каждый срез представляет собой пропорцию, а сумма каждый пропорции (среза) составляет 1 (или 100%)

Эти срезы предназначены для того, чтобы предоставить зрителям относительный размер данных, подобно тому, как действуют столбцы в столбчатой диаграмме. Длина дуги (размер угла срезы) каждого среза пропорциональна кол-ву, которое он представляет. Это также означает, что площадь каждого среда также пропорциональна этому кол-ву.

Прежде чем мы погрузимся в варианты использования круговых диаграмм и способы их построения, мы хотим рассмотреть тот факт, что круговые диаграммы являются наиболее проблемным графическим визуальным представлением. Они по прежнему являются одним из наиболее часто используемых инструментов визуализации, поэтому изучеие их будет жизненно важным для аналитика данных.

Основная причина использования круговой диаграммы 0 представление отношения "часть-целое" в ваших данных. Если вы хотите сравнить относительные размеры ваших данных и увидеть, как все они являются частью большей суммы, не беспокоясь об их точных размерах, круговая диаграмма может быть хорошим вариантом.

## Особенности круговых диаграмм

### Сравнение размеров категорий

Один из основных моменто визуализации данных является облегчение для читателя сравнения различных наборов данных информации. Срезы на круговой диаграмме треудно сранивать, посколько человеческий глаз с трудом сравнивает области. Давайте рассмотрим пример ниже

![](./img/2_9.png)

Гляда на любую из этих круговых диаграмм, вам будет трудно воспринять относительные размеры каждого среза. Если нашщей аудитории нужно это тонкое понимание данных, то столбчатая диаграмма будет гораздо эффективнее.

![](./img/2_10.png)

Давайте посмотрим на теже данные, представленные в виде столбчатых диаграмм.

Вы можете сразу определить относительный размер каждого столбца здесь. Это был бы тот случай, когда нам определенно следует отказать от круговой диаграммы.

````python
# для выполнения работы воспользуйте набором данных из файла ./files/pie_data.csv

# загрузим данные 
pie_data = pd.read_csv('./files/pie_data.csv')

# выведем содержимое файла
print(pie_data)

# создадим две переменные
graph_counts = pie_data['values']
graph_labels = pie_data['labels']

# постройте подграфик слева который будет являтся круговой диаграммой
plt.subplot(1,2,1)
plt.pie(graph_counts, labels=graph_labels)
plt.axis('equal')
plt.title('Tough to Compare')

# постройте подграфик справа который будет является столбчатой диаграммой
plt.subplot(1,2,2)
sns.barplot(pie_data, x=graph_labels, y=graph_counts)
plt.title('Easy to Compare')

plt.show()
````

## Улучшаем круговую диаграмму

Хороший жизненный девиз обычно звучит так: "чем больше кусков пирога, тем лучше". К сожалению, это не относится к круговым диграммам. При слишком большом кол-ве кусков попытка расшифровать визуальную часть становиться громоздкой - иногда даже более громоздкой, чем чтение данных из таблицы. Взгляните на пример ниже. На графике двадцать клиньев. ПОпытка сопоставить каждый кусок с его меткой - слишком много усилий, когда визуальная часть должна сделать читаемость простой для зрителя.

![](./img/2_11.png)

Когда вы оказываете в подобной ситуации, у вас есть несколько вариантов
- вы можете объединить части пирога, чтобы их было меньше, но при этом история оставалась информативной
- вы можете увидеть, выполняет ли столбчатая диаграмма более эффективную работу по отображению информации. Столбчатые диаграммы не страдают так сильно, как круговые диаграммы, когда дело доходит до визуализации многочисленных категорий. Это потому, что они не ограничены одним кругом - вместо этого они имеют многочисленные полосы

````python
# Для работы воспользуйтесь набором данных в файле ./files/major_data.csv и ./files/major_data_agg.csv
major_data = pd.read_csv("./files/major_data.csv")
print(major_data.head())

major_data_agg = pd.read_csv("./files/major_data_agg.csv")
print(major_data_agg.head())

pie_wedges = major_data["proportion"]
pie_labels = major_data["major"]

pie_wedges_agg = major_data_agg["proportion"]
pie_labels_agg = major_data_agg["department"]

plt.subplot(2,1,1)
plt.pie(pie_wedges, labels = pie_labels)
plt.axis('Equal')
plt.title("Too Many Slices")
plt.tight_layout()

plt.subplot(2,1,2)
plt.pie(pie_wedges_agg, labels=pie_labels_agg)
plt.axis('Equal')
plt.title("Good Number of Slices")
plt.tight_layout()

plt.show()
````
