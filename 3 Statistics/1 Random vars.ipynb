# Случайные величины

Случайная величина, в простом понимании это функция. В вероятности, мы часто используем случайные величины для представляения случайного события. Например, мы можем использовать случайную величину для представления результата броска игрального кубика: любое целое число от 1 до 6.

Случайными величинами должны быть числа, это значит что лучше использовать числа чем строки с характеристиками. Если мы хотим использовать случайные величины для представления события не с числовым результатом, мы можем выбрать числа для представления данного результата. Например, мы можем представить результат броска монетки случайной величиной как "орел" это 1 а "решка" 0.

В данном модуле мы покажем как использовать метод библиотеки numpy **`random.choise(a, size=size, replace = True/False)`** для симуляции случайных величин в python.

Данный метод принимает три переменные:
- *а* это список или объект который содержит значения из которого будет производится выбор
- *size* это число которое представляет как много значений следует выбрать
- *replace* может быть равен True или False, и обозначает оставляем ли мы выбранное значение в объекте *a* (replace = True) или удаляем его из объекта *a* (replace = False)

```python
# напишите код который будет симулировать бросок игральной кости используя модуль случайных величин
import numpy as np

# формируем объект значений
dice = range(1, 7)

# используя метод random симулируем бросок игральной кости 10 раз
rolls_results = np.random.choice(dice, size=10, replace=True)

# выводим результат
print(rolls_results)
```

## Discrete и Continuous случайные величины

### Discrete случайные величины

Случайная величина с посчитанным числом возможных значений. Напимер, бросок обычной игральной кости может иметь только 6 возможных значений.

Discrete случайная величина также используется когда наблюдатель может подсчитать события, например как кол-во людей вошедших в магазин в случайно выбраный день. В данном случае, значение можно подсчитать целыми числами.

### Continuous случайные величины

Когда возможные значенния случайной переменной невозможно подсчитать. В основном это переменные которые измеряют и они являются неподсчитываемыми по причине того что измерения могут всегда иметь точность измерения.

Например температура в Лос-Анджелесе в случайно выбранный день это continuous случайная величина. Мы всегда можем быть более точными до целых, десятый или даже сотых долей.
