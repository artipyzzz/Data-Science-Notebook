Но мы можете быть уверены что любая выборка из 30 песен вашего плейлиста подойдет к вечеринке. Для того чтобы сделать это, вы можете использовать силу распределения выборки.

```python
# импортируем библиотеки необходимые для работы
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import helper_function
```

```python
# Загрзите данные в переменную
spotify_data = pd.read_csv('../files/spotify_data.csv')

# выведите первые пару строк данных
print(spotify_data.head())
```

```python
# создайте переменную song_tempos и присвойте ему значение из набора данных одноименной колонки
song_tempos = spotify_data.tempo
```

```python
# используя вспомогательные методы выведите график распределения популяции
helper_function.population_distribution(song_tempos)
```

```python
# постройте диаграмму распределения выборки средних значений
# размер выборки 30
helper_function.sampling_distribution(song_tempos, 30, 'Mean')
```

```python
# постройте график распределения выборки минимальных значений
# размер выборки 30
helper_function.sampling_distribution(song_tempos, 30, 'Minimum')
```

```python
# постройте диаграмму распределения выборки показателя вариации
# размер выборки 30
helper_function.sampling_distribution(song_tempos, 30, 'Variance')
```

```python
# Рассчитайте показатели среднего значения и стандартного отклонения
# Запишите показатели в переменные population_mean , population_std
population_mean = song_tempos.mean()
population_std = song_tempos.std()
```

```python
# расчитате показатель стандартной ошибки
# с размером выборки 30
standard_error = population_std / (30 ** .5)
```

```python
# рассчитайте вероятность что средний темп 30 случайных песен будет ниже 140 bpm
print(stats.norm.cdf(140, population_mean, standard_error))
```

```python
# рассчитайте вероятность что средний темп 30 случайных песен будет больше 150 bpm
print(1 - stats.norm.cdf(150, population_mean, standard_error))
```
