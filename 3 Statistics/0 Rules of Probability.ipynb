# Rules of Probability (Правила вероятности)

## Union, Intersection and Complement (Объединение, Пересечение и Исключение)

### Union (Объединение)

Объединение двух множеств составляется другое множество в котором каждое из множеств составивших его являются подмножествами

![Union](./img/0_1.png)

Например, давайте предположим что у нас есть два множества, *A* и *B*. *A* представляет собой нечетные числа шести гранного игрального кубика ({1,3,5}). *B* представляет четные числа шести гранного игрального кубика. Объединение двух этих множеств является множество из элементов каждого из них ({1,2,3,4,5,6})

Математически мы можем представить объединение как *(A or B)*

### Intersection (Пересечение)

Пересечение двух множеств является множество которое представляет собой подмножество двух этих множеств.

![Intersection](./img/0_2.png)

Предположим что множество *A* представляет собой нечетные числа шести гранного игрального кубика. А множество *B* числа от 2 до 6 (или другими словами больше 2)

Таким образом пересечение будет множество состоящее из элементов {3,5}

Математически пересечение представляется так *(A and B)*

### Complement (Исключение)

Исключение множества представляет собой все возможные значения другого не входящие в него.

![Complement](./img/0_3.png)

Другими словами если множество *A* представляет собой четные значения шести гранного игрального кубика в таком случае исключением множества *А* является множество состоящее из значений {2,4,6}

Математически исключение множества *А* можно представить в виде *$A^C$*

## Independence and Dependence (Независимость и зависимость)

Представьте что вы подбрасываете монетку 5 раз и получаете 5 орлов подряд. Повлияет ли это на результат того что 6 раз вам также выпадет орел? Не смотря на все мы будет чувствовать что сейчас тот самый момент когда выпадет решка, невозможно чтобы предыдущие подбросы монет повлияли на будущие. Факт того что предыдущие броски не влияют на результат следующих называют **Независимостью** (**Independence**). Два события являются независимыми если появление одного события не влияет на вероятность другого.

Представим что у нас есть мешок с шариками кол-во которых составляет 5 штук
- 2 синих
- 3 красных

Если мы возьмем один шарик из мешка, каков шанс того что второй шарик будет синим?

![Dependence](./img/0_4.png)

Изображение выше объясняет явление **Зависимости (Dependece)** событий. Вероятность взять синий шарик вторым зависит от того взяли ли мы в первый раз красный или синий шарик.

Что произойдет если мы положим первый шарик который взяли обратно? Будет ли вероятность взять синий шарик вторым зависим или независимым в зависимости от того какой шарик мы взяли первым? В таком случае событие являтся независимым.

![Independence](./img/0_5.png)

Почему нас беспокоит является ли событие зависимым или независимым? Эти знаний помогут нам расчитать веростноять события. Это помогает исследователям понимать и предсказывать результаты комплексных процессов таких как:
- эффективность вакцины
- погода в определенный день

## Взаимоисключающие события (Mutually Exclusive Events)

Два события являются взаимоисключающими в случае если они не могут возникнуть одновременно. Например, вернемся к подбросу монетки: событие "решка" и "орла" взаимоисключающие события потому как мы не можем одновременно получить и "решка" и "орел" одновременно за один бросок.

Визуально мы можем представить два взаимоисключающих события как пару не перекрывающихся круга. Она не перекрывают друг друга потому что не имеют общего результата который бы удовлетворял и первое и второму событию.

![Exclusive](./img/0_6.png)

Что мы можем сказать о взаимоисключающих событиях? Если событие *A* включается в себя нечетные числа игрового кубика и событие *B* это множество значений игрального кубика больше 2, в таком случае события не взаимоисключающие. У них есть пересечение элементов множеств ({3,5}). Любое событие которое имеет не пустое множество пересечения двух множеств является не взаимоисключающим.

## Правило добавления (Addition Rule)

Давайте теперь разберем концепцию расчета вероятности

Вернем к одному из наших первых примеров: событие *A* множество из значеий нечетных чисел игрального кубика, событие *B* множество значений больше 2 игрального кубика. Что если мы хотим получить вероятность одного или двух событий сразу? Это называют вероятность объединения событий *A* и *B*

`P(A or B) = P(A) + P(B) - P(A and B)`

Мы вычитаем пересечение событий *A* и *B* потому что оно включается дважды в сложении вероятности события *A* и *B* (P(A), P(B))

Что если события взаимоисключающие? За один бросок игральной кости, если событие *A* это множество значений состоящий из элементов меньше 2 и событие *B* множество значений больше 5 игральной кости, в таком случае оба события не могут произойти одновременно.

`P(A or B)=P(A) + P(B)`

Данная формула справедлива так как пересечением является пустым множеством.

## Условная вероятность (Conditional Probability)

Если мы хотим рассчитать вероятность того что пара событий возникнет одновременно, нам необходимо определить условную вероятность. Используя мешок с шариками как пример, давайте вспомним определение зависимых событий.

![Dependence](./img/0_4.png)

Если мы возьмем два шарика из мешка в котором 5 шариков без возврата их обратно в мешок, вероятность того что второй шарик будет красным зависит от того какой цвет первого шарика. Для данной ситуации существует отдельный термин: **Условная вероятность**. В кратце, условная вероятность измеряет вероятность одного события, с учетом того что предыдущее событие уже произошло.

Было принято что условная вероятность записывается как pipeline (|). Например, если мы хотим получить вероятность того что мы получим красный шарик вторым учитывая что первый был синий можно записать так:

`P(Red Second | Blue First)`

Судя по диаграмме выше мы знаем что:

`P(Red Second | Blue First) = 3/4`

Что если мы взяли два шарика и вернули бы их? Как бы выглядела условная вероятность? Давайте подумаем об этом. Вне зависимости от того какой шарик мы взяли первым, он будет возвращен обратно в мешок. В таком случае вероятность вытащит красный шарик или синий шарик вторым независимо от того какой шарик был взят первым.

И это мы можем представить в виде:

`P(A|B)=P(A)`

`P(B|A)=P(B)`

## Правило перемножения (Multiplication Rule)

Раннее мы рассматривали правило добавления, которое описывает вероятность возникновения одного ИЛИ другого события (или оба события). Что если мы хотим рассчитать вероятность того что два события произойдет одновременно. Для двух событий, *A* и *B*, вероятность описывается так **P(A and B)** или другими словами вероятность множества пересечения множеств A и B.

### Основная формула

Основная формула расчета вероятности что два события произойдет одновременно такова:

`P(A and B) = P(A) * P(B|A)`

Но мы можем сократить данную формулу для независимых событий.

### Зависимые события

Давайте вернемся к нашему примеру - мешку с шариками. У нас есть пять шариков: 2 синих и 3 красных. Мы берем два шарика без возврата их обратно в мешок. Что если мы хотим узнать вероятность того что первый шарик будет синий и второй шарик также будет синим?

Принимая во внимание условную вероятность, правило умножения для этих двух зависимых событий имеет вид:

`P(Blue 1st and Blue 2nd) = P(Blue 1st) * P(Blue 2nd | Blue 1st)`

`P(Blue 1st and Blue 2nd) = 2/5 * 1/4`

`P(Blue 1st and Blue 2nd) = 1/10`

Это один из потенциальных результатов при взятии двух шариков из мешка. Один из способов визуализировать все возможные результаты пары событий это диаграмма дерева (**Tree Diagram**)

Древовидная диаграмма имеет следующие свойства:
- Каждая ветка представляет собой некоторое множество событий
- Вероятность родительских веток (все возможные результаты события) в сумме дает 1
- Мы перемножаем значения сквозь ветки (используя правила умножения) для расчета вероятности каждой ветки (множества результатов) которые могут возникнуть.

![Multiply](./img/0_7.png)

### Независимые события

Для двух независимых событий, правило перемножения становится менее сложной. Расчет вероятности двух независимых событий представляется формулой

`P(A and B) = P(A) * P(B)`

Это происходит потому что формула условной вероятности для независимых событий принимает другой вид:

`P(B|A) = P(B)`

Давайте взглянем на простой пример: подбросим монетку дважды. Событие *A* заключается в том что мы получим решку при первом подбросе, и событие *B* говорит о том что мы получим решку на втором подбросе. `P(A=P(B)=.5`, таким образом следуя формуле выше, вероятность получить две решки на каждый подброс будет равна:

`P(A and B) = .5 * .5 = .25`

Визуально такой случай можно представить в виду древовидной диаграммы

!["Coin flip Tree Diagram"](./img/0_8.png)

## Условная вероятность (Продолжение)

Мы представили условную вероятность как часть правила перемножения для зависимых событий. В любой случае, давайте немного погрузимся в данную тему и узнаем насколько это хороший инструмент вычисления вероятности которое нашло свое применение в реальном мире.

Для проблемы описанной ниже будет использоваться следующая диаграмма

!["Diagram conditional probability"](./img/0_9.png)

Предположим что следующее предположение правдиво (то что отображенно в первом множестве веток диаграммы)
- 20 процентов популяции болеют ангиной
- 80 процентов популяции не болеют ангиной

Теперь предположим мы проводим тест по ветке популяции которая болеет ангиной. Возможные результаты данного теста отображены далее в множестве веток
- Если человек более ангиной, то существует 85% вероятность что его тест на болезнь будет положительный и 15% шанс что он будет отрицательным. Это можно обозначить как:

`P(+|ST)=.85`

`P(-|ST)=.15`

- Если человек не болеет ангиной то существует вероятность 98% что его тест будет отрицательным и 2% шанса что он будет положительным. Это можно обозначить как:

`P(-|NO ST)=.98`

`P(+|NO ST)=.02`

- Наконец, давайте посмотрим на 4 возможные пары результатов которые формируются от корневых веток нашей диаграммы:

`P(ST and +)=.17`

`P(ST and -)=.03`

`P(NO ST and +)=.016`

`P(NO ST and -)=.784`

Сумма всех значеий составляет единицу, поскольку охватывают все возможные результаты после тестирования пациентов.

Это отлично что у нас есть вся эта информация. В любом случае, мы что то упустили. Если кто-то получить положительный результат, какова вероятность того что он более ангиной? Условно, мы можем записать данную вероятность как:

`P(ST|+)`

## Теорема Байеса (Bayes' Theoreme)

Представьте что вы пациент который недавно был положительно оценнен как болеющий ангиной. Вы хотели бы узнать какова вероятность того что вы на самом деле болеете ангиной на основании что ваш тест положительный

`P(ST | +)`

Для расчета вероятности, мы будет использовать нечто что называют Теоремой Байеса, которая рассчитывается по формуле

`P(B|A)=(P(A|B)*P(B))/P(A)`

Используя теорему Байеса

`P(ST|+)=(P(+|ST) * P(ST))/P(+)`

Мы знаем что

`P(+|ST) = .85`

Также мы знаем что

`P(ST)=.20`

Что насчет **P(+)**? Это то что мы знаем? Давайте подумаем об этом. Существует четыре возможных результата:
- Болеет ангиной и тест положительный
- Болеет ангиной и тест отрицательный
- Не более ангиной и тест положительный
- Не более ангиной и тест отрицательный

Нам интересны только два результата когда у пациента положительный результат **P(+)**. Таким образом мы можем сказать что

`P(+)=P(ST and +) + P(NO ST and +)`

`P(+)=.17 + .016`

`P(+)=.186`

Наконец подставив все значения в формулу теоремы Байеса мы получим

`P(ST|+)=(.85 * .20)/.186 = .914`

Существует 91.4% процент вероятности того что вы на самом деле болеете ангиной в случае если получили положительный результат проверки. Это не очевидно из информации изложенной в нашей древовидной диаграмме, но с помощью теоремы Байеса мы смогли это вычислить.
